Import Libraries

```python
!pip install opencv-contrib-python
import cv2
import numpy as np
import matplotlib.pyplot as plt
!pip install Pillow

%matplotlib inline
```

Create a black image to work


```python
black_img = np.zeros(shape= (512, 512, 3), dtype=np.int16)
```

Get the shape of the image

```python
black_img.shape
```

Show It

```python
plt.imshow(black_img)
```

Draw a cricle

```python
cv2.circle(img=black_img, center=(400,100), radius=50, color=(255, 0, 0), thickness=8)
plt.imshow(black_img)
```

Filled Circle


```python
cv2.circle(img=black_img, center=(400,200), radius=50, color=(0, 255, 0), thickness=-1)
plt.imshow(black_img)
```

Draw Rectangle

```python
cv2.rectangle(black_img, pt1=(200,200), pt2=(300,300), color=(0,255,0), thickness=6)
plt.imshow(black_img)
```

Draw Triangle

```python

vertices = np.array([[10, 450], [110, 350], [180, 450]], np.int32)

pts = vertices.reshape(-1, 1, 2)

cv2.polylines(black_img, [pts], isClosed=True, color=(0, 0, 255), thickness=3)

plt.imshow(black_img)
```

Draw Filled Rectangle

```python
cv2.rectangle(black_img, pt1=(300,300), pt2=(400,400), color=(110,200,255), thickness=-1)
plt.imshow(black_img)
```

Filled Triangle

```python
vertices = np.array([[180, 450], [250, 350], [300, 450]], np.int32)

pts2 = vertices.reshape(-1, 1, 2)

cv2.fillPoly(black_img, [pts2], color=(255, 200, 110))

plt.imshow(black_img)
```

Draw Line

```python
cv2.line(black_img, pt1=(512,0),pt2=(0,512), color=(200, 200, 200), thickness=7)
plt.imshow(black_img)
```

Write Text

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(black_img, text='Rajesh', org=(210, 500), fontFace=font, fontScale=3, color=(255, 255, 0), thickness=3, lineType=cv2.LINE_AA)
plt.imshow(black_img)
```
